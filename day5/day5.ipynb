### how does passenger survival rate depend on gender?

```python
df.groupby('Sex')[['Survived']].aggregate(lambda x:x.sum()/len(x)).plot(kind='bar')
```

```python
df.groupby('Pclass')[['Survived']].aggregate(lambda x:x.sum()/len(x)).plot(kind='bar')
```
