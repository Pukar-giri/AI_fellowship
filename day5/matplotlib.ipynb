```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
```

```python
x=np.arange(6,20)
y=np.random.rand(14)
assert(x.shape==y.shape)
```

```python
linregressor=LinearRegression(fit_intercept=True,normalize=False,n_jobs=-1)
linregressor.fit(x.reshape(-1,1),y)
```

```python
y_pred=linregressor.predict(x.reshape(-1,1))
```

```python
plt.plot(x,y_pred,color="red")
plt.scatter(x,y,color='blue',marker='v')
```
