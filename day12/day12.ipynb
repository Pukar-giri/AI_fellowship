```python
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
```

```python
w1=np.linspace(-2,2,500)
w2=np.linspace(-2,2,500)
w1,w2=np.meshgrid(w1,w2)
```

```python
w1
```

```python
w2
```

```python
j=np.square(w1)+np.square(w2)+5
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
ax.contour3D(w1,w2,j,500)
```
